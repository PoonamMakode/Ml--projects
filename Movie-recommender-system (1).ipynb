```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# check genres_ as same apply  convert method on keywords ,cost, crew
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# i want only real name of actor not caractor_name of actor so what we do _

def main_actor(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!= 2:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l

```

```python
movies['cast']=movies['cast'].apply(main_actor)
```

```python
movies.head()
```

```python
# only i want director name_ so what we do- lets see

def fetch_director(obj):
    l =[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'] =movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# remove space btw the names it's vry imp for recommands system for get accurate recommand result

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()

# see here name nd last name space has removed
```

```python
# concatinate
movies['tags']=movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
new_df=movies[['movie_id','title', 'tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# steming methode for use== like, likes , liked == like ,like ,like

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
        
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
# similarity btw two movies in 5000 movies...convert vector
# for vactorization we use sklearn libraries
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
# find out the distance btw one movie to another , if the distance is small then the simlarities btw movies less...
# we can't measure uclidient distance...it's not good for high dimensional data
# cosine distance is a angle distance.....
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
# store in similarity variable
similarity = cosine_similarity(vectors)
```

```python
# list(enumerate(similarity[0]))

# index with position
```

```python
# i want top 5 movies
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
# similar 5 movies return

def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distance= similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
